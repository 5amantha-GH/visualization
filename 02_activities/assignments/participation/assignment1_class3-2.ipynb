```python
# loading our libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

# creating our sample data
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])   # new categorical data
y2 = np.array([110, 180, 240, 99, 220])
```

```python
# creating subplots
fig, (ax1, ax2) = plt.subplots(ncols=2, 
                               nrows=1, 
                               figsize=(7, 3))

ax1.scatter(x1,y1)
ax2.bar(x2,y2)
```

```python
# activity: customizing our plots
fig, (ax1, ax2) = plt.subplots(ncols=2,
                        nrows=1,
                        figsize=(9, 3))

font1 = {'family':'serif','color':'indigo', 'size':10}
font2 = {'family':'serif','color':'indigo', 'size':14}

ax1.scatter(x1,y1,
            marker ='*',
            color = 'red',)

ax2.bar(x2,y2, color = 'orange')

ax1.set_title('Scatter Plot', fontdict=font2)
ax2.set_title('Pirate Strength', fontdict=font2)
ax1.set_xlabel('Time', fontdict=font1)
ax2.set_xlabel('Pirate Crew Member', fontdict=font1)
ax1.set_ylabel('Measure', fontdict=font1)
ax2.set_ylabel('Strength', fontdict=font1)
```

```python
# arranging our subplots
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],    # changes how much each plot takes up the figure space
                                    figsize=(7, 4))

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
```

```python
# adding axes labels to our subplots in very large font size
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize=(7, 4))

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)     # this label is completely cut off!
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18)
```

```python
# using layout to make sure our subplots fit neatly in our figure area
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize=(7, 4),
                                    layout = "constrained")

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label', fontsize=18)
someaxes["ax2"].set_xlabel('Another Label', fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)
```

```python
# multiple visualizations on one axes object
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

# define our figure and axes (just one this time)
fig, ax = plt.subplots(figsize=(7, 3))

# now call both bar and plot elements to the same axes (ax)
ax.bar(x, y1, color = "indigo")
ax.plot(x, y2, color = "red")
```

```python
# adding error info plotted such as std dev, range, etc.

# calculate standard deviation of our data
y2_sd = np.std(y2)

# then plot our line as before
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

# specifying our x and y-values to know where were putting our error bars
ax.errorbar(x, # our x values
            y2, # our y values
            yerr = y2_sd,
            fmt = "none")   # for visual reasons, we want to define our line separately this way (fmt = format)
```

```python
# customizing errorbar appearance
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

ax.errorbar(x, # our x values
            y2, # our y values
            yerr = y2_sd,
            fmt = "none",   
            ecolor = 'indigo',
            elinewidth = 4,
            capsize= 6,
            capthick = 4)
```

```python
# if we dont want to see error bars for every single point, we can specify intervals using errorevery
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

ax.errorbar(x, # our x values
            y2, # our y values
            yerr = y2_sd,
            fmt = "none",   
            ecolor = 'indigo',
            elinewidth = 4,
            capsize= 6,
            capthick = 4,
            errorevery= 2)
```

```python
# using images from the internet
# first load our libraries
from PIL import Image   # to open images
import requests         # to get images from URLs
from io import BytesIO  # to store images
```

```python
# then get our image from the internet
response = requests.get("https://static.wikia.nocookie.net/the-ultimate-good/images/3/3d/Luffy.png/revision/latest?cb=20230417200546")
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
# make a basic line plot
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

# then overlay a new axis (ax_image) on our figure (on top of ax) to act as a container for our image
ax_image = fig.add_axes([0.1, # x coordinate (ON FIGURE, NOT AXES)
                         0.11, # y coordinate (ON FIGURE, NOT AXES)
                         0.15, # image width
                         0.35]) # image height

# then add imshow() to add the image we prepared before
ax_image.imshow(image)
ax_image.axis('off')
```
