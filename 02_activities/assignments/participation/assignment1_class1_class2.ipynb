```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# generating our sample data
np.random.seed(613)                 # set random seed for reproducibility
x = np.arange(50)                   # gives us a range of 50 values
y = np.random.randint(0,100,50)     # gives us 50 random integers between 0-100

# make a figure (our container/canvas) and within that figure we want to put an axis/plot object called 'ax'
fig, ax = plt.subplots(figsize=(5,3))   # setting the size of our figure to 5 x 3
ax.scatter(x,y)                         # makes scatter plot for our x and y-variables that we defined earlier
```

```python
# vertical bar chart
fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

```python
# horizontal bar chart
fig, ax = plt.subplots(figsize=(5,3))
ax.barh(x,y)
```

```python
# line plot
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

```python
# histogram
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y) 

# only looks at y-variable because histograms count up the number of occurrences of something
# so the y-axis here is counting up how many times each of the values in our x-axis occurs
```

```python
# adding axis titles to our line plot
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time')  # title for our plot
ax.set_ylabel('Total growth')   # x-axis label
ax.set_xlabel('Years since start')  # y-axis label
```

```python
# changing how our labels and titles look using font dictionaries
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

# apply the appropriate font dictionaries to our labels/titles
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
# moving the alignment of our lables/titles with the 'loc' argument
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict = font1, loc = 'left')      # movement occurs here
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
# changing the appearance of our data points using the 'marker' argument
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter( x,y, 
           marker = '*',            # many different marker shapes exist
           color = "indigo")
```

```python
# adding a line style that connects to our points
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot( x,y,                           # chart style changed from scatter to plot (ax.plot) to see the connecting lines
           marker='*', 
           color = "indigo",
           linestyle = '--',
           linewidth = 2)
```

```python
# changing marker size and marker colours using hex codes
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot( x,y, 
           marker='*',
           markersize = 12, 
           color = "indigo",
           linestyle = '--',
           linewidth = 2,
           markeredgecolor = '#fa9359',
           markerfacecolor = '#000000')
```

```python
# adding gridlines to our axis object
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot( x,y, 
           marker='*',
           markersize = 12, 
           color = "indigo",
           linestyle = '--',
           linewidth = 2,
           markeredgecolor = '#fa9359',
           markerfacecolor = '#000000')

ax.grid(axis = 'y')
# y - creates horizontal gridlines, x - creates vertical gridlines, leaving blank creates both horizontal and vertical gridlines
```

```python
# changing gridline appearance
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot( x,y, 
           marker='*',
           markersize = 12, 
           color = "indigo",
           linestyle = '--',
           linewidth = 2,
           markeredgecolor = '#fa9359',
           markerfacecolor = '#000000')

ax.grid(axis = 'y',
        color = "black",
        linestyle = '-.',
        linewidth = 0.5)
```

```python
# activity from https://python-graph-gallery.com/scatter-plot/ with subtle self-made changes

# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# set plot style: grey grid in the background:
sns.set_theme(style="darkgrid")

# load dataset
tips = sns.load_dataset("tips")

# Create an array with the colors you want to use
colors = ["#69b3a2", "#4374B3"]
sns.set_palette(sns.color_palette(colors))

# Set the figure size
plt.figure(figsize=(7, 7))

# grouped barplot
ax = sns.barplot(
    x="day", 
    y="total_bill", 
    hue="smoker", 
    data=tips, 
    ci=None
    )

# Customize the axes and title
ax.set_title("Smokers have bigger bills")
ax.set_ylabel("Bill value")
ax.set_xlabel("Day")

# Remove top and right borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```
