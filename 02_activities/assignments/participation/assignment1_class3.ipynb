```python
# load our libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

# create our sample data
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

```python
# activity: plot both y-variables on the same axes
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1)
ax.plot(x,y2)

# we plot the y-variables separately so we can modify the appearance of the lines differently
```

```python
# adding a legend
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc = 'lower right')
```

```python
# modifying the appearance of our legend
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc = 'lower right',
          frameon = True,   # add frame around the legend
          fontsize = 12,    # change font size
          ncol = 2,         # specify number of columns
          shadow = True)    # adds a dropshadow frame to the legend
```

```python
# moving legend outside of plot area
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc = 'upper left',
          bbox_to_anchor =(1, 1))   # 1,1 are the coordinates of our figure
```

```python
# adding annotations with the text() function
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = 'lower right')

ax.text(10, 95, "This value is important!")   # 10,95 are the coordinates of a data point we've chosen to highlight
```

```python
# modifying the appearance of text annotations
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = 'lower right')

ax.text(10, 95, "This value is important!",
        ha='center',    # alignment of text
        color = 'red',  # modify font colour
        size = 20)      # modify font size
```

```python
# positioning text annotations using transforms
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])     # create an empty set of axes, 0-10 horizontally and 0-10 vertically

# how were specifying our coordinates based on location
ax.text(1, 5, ". Data:(1, 5)", transform=ax.transData)              # transData - transforms to our data
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform=ax.transAxes)      # transAxes - tranforms to our axes object
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform=fig.transFigure) # transFigure - transforms relative to our figure
```

```python
# annotating with arrows
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),   # coordinates are where we want our arrow and text to be
            arrowprops=dict(facecolor='black'))
```

```python
# modifying arrows - https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.annotate.html
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),
            arrowprops = dict(facecolor = 'hotpink', arrowstyle = 'fancy'))
```

```python
# removing tick marks and labels on our axes, more used for qualitative visualizations
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
# defining a max number of ticks on our axis
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
# setting tick marks in intervals
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.xaxis.set_major_locator(plt.MultipleLocator(5))  # intervals of 5
```

```python
# rotate the text of our tick labels
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

plt.xticks(rotation=45, ha='right', fontsize=12, font='serif')
```

```python
# modifying our axis title font
font1 = {'family':'serif','color':'indigo'}

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

plt.xticks(rotation=45, ha='right', fontsize=12, font='serif')
plt.xlabel('Shiny New X Axis!', fontsize = 18, fontdict = font1)
```
